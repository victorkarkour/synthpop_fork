```python
datafile = pandas.read_csv("/home/victor/SURP/outputfiles/hst_bulge_treasury_sim/smalldeg/Oct2024_l-10.000_b-6.000.csv")
BP = datafile["Gaia_BP_EDR3"]
VP = datafile["Gaia_RP_EDR3"]
BPVP = BP - VP

datafile.insert(len(datafile.columns),"BPVP", BP - VP)
```

```python
Cond = (datafile.BPVP > 0.95) & (datafile["Gaia_G_EDR3"] < 3.9)


dataCond= datafile.loc[Cond, ["Dist", "VR_LSR","pop"]]
dataCondB = dataCond.loc[dataCond["pop"] == 0, ["Dist", "VR_LSR"]]
dataCondD = dataCond.loc[dataCond["pop"] != 0, ["Dist", "VR_LSR"]]

numlist = []
binlist = []
ylist = np.linspace(0, 25)
j = -500
while j <= 500:
    RV = dataCond.loc[(dataCond["VR_LSR"] >= j-25) & (dataCond["VR_LSR"] <= j+25), ["VR_LSR"]]
    numlist.append(len(RV))
    binlist.append(j)
    j += 25

# Gaussian from paper
g_init = models.Gaussian1D(amplitude = 11.5, mean = -86, stddev= 72)

# print(numlist)
fig, ax = plt.subplots()
ax.scatter(binlist, numlist, label = "SynthPop Data")
ax.plot(binlist, g_init(binlist), color = "red", label = "Paper Fit")
ax.set_ylabel("Counts")
ax.set_xlabel("RV LSR")
ax.set_title("Velocity Dispersion for (-10 l,-6 b)")
ax.text(200,14, "Mean = 47.10 [km/s]")
ax.text(200,12.5, "$\sigma$ = 103.22 [km/s]")
plt.legend()
plt.show()
```
